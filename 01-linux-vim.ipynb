## Linguagem de programação e compilador: 

Os números, e caracteres em geral, podem ser escritos em diferentes bases:  
  
  
0001(2) = 1(10)  
0010(2) = 2(10)  
0100(2) = 4(10)  
1000(2) = 8(10)  
  
Logo: 1111 (2) 	= 8(10) + 4(10) + 2(10) + 1(10) = 15 (10)  

O processador de um computador realiza operações binarias, ou seja, com números e caracteres representados por 0's e 1's.  Porém os números em binario podem ser muito longos: 1001111000001010. Assim, normalmente a linguagem de máquina é a base hexadecimal, ou hex, pois convertir binario a hex é mais simples que converter de binario a decimal, exemplo:  
  
  01011110101101010010(2) 	= 262144(10) + 65536(10) + 32768(10) + 16384(10) + 8192(10) + 2048(10) + 512(10) + 256(10) + 64(10) + 16(10) + 2(10)   
  = 387922(10)  
  
  01011110101101010010(2) = 0101  	1110  	1011  	0101  	0010 (2)  
    = 5 	E 	B 	5 	2 (16)  
    = 5EB52 (16)

## Breve historia do Linux

[1972](https://en.wikipedia.org/wiki/C_(programming_language)) - Surgimento da linguagem C (Ken Thompson e Dennis Ritchie, que trabalhavam na Bell Labs). Diferentes empresas criaram compiladores, proprietarios, de C  
[1987](https://en.wikipedia.org/wiki/GNU_Compiler_Collection) - O compilador de C, livre, gratuito, e de código aberto,  GNU,  foi desenvolvido pela Free Software Foundation, criada por Richard Stallman, entre outros  
[1991](https://en.wikipedia.org/wiki/History_of_Linux) - Linus Torvalds desenvolve o Kernel para controlar um processador 80386. Para executar o código, escrito em C, no processador,  Linus usou o compilador GNU. Já que um Kernel não serve de nada sem um compilador e uma terminal e muitas bibliotecas, Linus publicou seu código usando a licencia GNU, a qual é usada no compilador e bibliotecas que ele usou.  A partir de então o sistema operativo se conhece como GNU-Linux.  
[2023](https://www.kernel.org/) - A versão estável mais recente do Kernel de Linux é a 6.4.12


Graças à licência GNU, existem centos de diferentes distribuições: Debian, Slackware, Red Hat, Ubuntu, Mint, etc ...

E uma versão mais familiar: ANDROID

Mesmo sendo pouco usado como desktop, onde Microsoft é ainda o preferido, Linux é a linguagem mais usada em servidores (inclusive os servidores de Microsoft funcionam baixo alguma versão do kernel GNU-Linux).  O sistema operacional de "smartphones", Android,  é baseado em Linux e existe em 75% dos telefones do mundo.  

Distribuição sugerida de Linux: [POP-OS](https://pop.system76.com/)

Windows disponibiliza a opção de rodar linux sem desinstalar o SO windows usando [WSL](https://learn.microsoft.com/pt-br/windows/wsl/install)

A melhor opção contudo, é fazer um "dual boot" fazendo uma partição do disco rígido.

## A terminal de Linux

Uma terminal é uma interface que interpreta comandos e "scripts" dos usuarios indicando ações ao computador. Ou seja, atraves da terminal o usuario disse ao computador o qué fazer.  Um excelente livro sobre como usar a terminal se encontra no seguinte link:

[The Linux Comand Line](http://linuxcommand.org/tlcl.php)


### Principais comandos de navegação

**pwd**: "Print Working Directory", mostra a localização na árbore do sistema  
  
  
**cd**: "Change Directory", para ir a outro diretorio  
  
  
**ls**: "List"  
  

### Trabalhando com arquivos e diretórios  
  
    
  
**file**: Encontra que tipo de arquivo é (e.g., **file /bin/ls** disse que "ls" é um arquivo executável)

**touch**: cria arquivos vazios (e.g., **touch aula1.txt** vai criar um arquivo sem conteúdo chamado aula1.txt)

**echo**: imprime na um texto entre aspas(e.g., **echo "olá mundo"**)
  
**cat**: Mostra o conteúdo de um arquivo na tela  
  
**head**: Mostra as últimas linhas de um arquivo de texto  
  
**tail**: Mostra as últimas linhas de um arquivo de texto, (e.g., **tail -f /var/log/messages**, mostra, de 5 em 5 linhas, as saidas do arquivo "messages"   
  
  
**cp**: copia um arquivo de um local para outro  
  
**mv**: move um arquivo de um local para outro ou muda o nome do arquivo  
  
**rm**: apaga um arquivo, para apagar um directorio e os arquivos contidos nele, deve se usar **rm -R" ou "rm -rf**  
  
**mkdir**: Cria um diretorio (e.g., **mkdir ./meu_nome**)  
  
**rmdir**: Apaga diretorio, similar a **rm -R**

**>  :** dirige um texto a um arquivo de saida (e.g., **echo "ola mundo" > aula1.txt**)

** >> :** adiciona texto a um arquivo de saida (e.g., **echo "este é meu primeiro arquivo criado usando a terminal" >> aula1.txt**)


### Encontrando coisas
  
**which**: Mostra o caminho completo a um comando (e.g., "which gfortran", a saida será: "/usr/bin/gfortran". 
  
**locate**: Forma rápida de encontrar arquivos em qualquer local do SO, o comando "locate cap01" indicará onde se encontrarm todos os arquivos que tenham a palavra cap01  
  
**find**: Serve para procurar por arquivos que tenham certos padrões, e.g., o comando "find . -name \*py" vai procurar no diretorio atual todos os arquivos com as palavras py no nome

**grep**: serve para procurar padrões de texto em documentos, e.g., o comando "grep 'print' \*.py" ira mostrar quais arquivos de python, dentro do diretorio atual possuem a instrução *print*.

# O editor vim

Já vimos como criar arquivos sem conteúdo.  Algo um pouco chato. Vamos aprender agora a utilizar um editor de texto que nos permite adicionar conteúdo, e edita-lo, dentro desses arquivos.  

Vim é uma versão do editor Vi com algumas funcionalidades extra. Porém, provavelmente nessa altura o programa Vi também seja desconhecido.  Esse tipo de editores é differente ao que pode ter sido usado até agora no computador.   

Como ou vim existem [outros](https://medium.com/linode-cube/emacs-nano-or-vim-choose-your-terminal-based-text-editor-wisely-8f3826c92a68) editores de texto que podem ser usados para editar documentos de LaTex, ou, e isso é o que mais nos interessa, escrever código de alguma linguagem de programação. Para nosso caso, o PYTHON. Vi, ou vim, é um dos editores de texto mais poderosos. Veremos porqué...

### Editor de linha de comando

O vim não tem entorno gráfico, ou seja, para editar em vim teremos necessariamente que abrir uma terminal (um editor similar em Windows é o *notepad* e em Mac OSx é o *Textedit*.  

Há dois modos, estados, nos quais o editor vim uma vez aberto pode estar:  

* **Edit**: é o modo *default* do vim. Nesse modo podemos navegar ao longo do arquivo utilizando as ferramentas que o vim tem para isso. Podemos apagar, copiar, procurar, substituir, salvar, etc.

* **Insert**: é o modo no qual é possível introduzir conteúdo no arquivo.

### Salvando e saindo

Há varias formas de fazer isto dependendo do que se quer fazer

* **ZZ** (em maiúsculas) - Salva e sai
* **:q!** - sai do vim sem salvar as mudanças feitas no arquivo
* **:w**  - salva as mudanças mas não sai do vim
* **:wq** - savla e sai

Como se ve, a maioria dos comandos precisa ser precedido pelos  *dois pontos*, **:**, logo deve ser apertada a tecla ENTER.

### Navegando um arquivo em vim

Teremos que estar no modo **Edit**. Existem varios comandos que nos permitem *navegar*:

* **Setas** - Movimentam o cursor a vontade
* **^** - Move o cursor ao inicio da linha
* **nG** - move o cursor à n-essima linha (eg. **5G** move o cursor a linha número 5)
* **:n** - similar a **nG**
* **G** - move o cursor à ultima linha
* **w** - move o cursor ao começo da proxima palavra (note que não precisa dos *dois pontos*, **:** nesse caso)
* **nw** - avança n palavras
* **b** - começo da palavra anterior
* **nb** - começo da n-essima palavra anterior
* **{** - retrocede um paragrafo
* **}** - avança um paragrafo

### Apagando counteúdo, copiando e colando

Alguns comandos para apagar conteudo, alem da tecla DELETE

* **x** - apaga um caracter
* **nx** - apaga n caracteres
* **dd** - apaga uma linha 
* **yy** - copia uma linha inteira
* **p** - cola uma linha que foi apagada com **dd** ou copiada com **yy*

### Procurando coisas

É possível navegar o documento a procura de palavras ou sequencias de caracteres. Por exemplo, se estamos interessados em procurar a palavra **plot**, comunmente usada no python para fazer uma figura, usamos

* **/plot** - procura a sequencia de caracteres **plot** para frente no documento
* **?plot** - procura para atras

### Desfazendo e refazendo

Caso cometamos um erro na edição do arquivo, podemos rapidamente desfazer esse erro, voltando ao modo **Edit** e apertando

* **u** desfaz a ultima mudança
* **U** desfaz todas as mudanças na linha do documento onde o cursor se encontra
